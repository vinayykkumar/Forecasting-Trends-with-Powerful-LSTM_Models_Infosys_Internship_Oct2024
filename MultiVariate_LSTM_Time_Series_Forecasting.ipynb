```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
temp = df['T (degC)']
temp.plot()
```

```python
# [[[1], [2], [3], [4], [5]]] [6]
# [[[2], [3], [4], [5], [6]]] [7]
# [[[3], [4], [5], [6], [7]]] [8]

def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
X_train1, y_train1 = X1[:60000], y1[:60000]
X_val1, y_val1 = X1[60000:65000], y1[60000:65000]
X_test1, y_test1 = X1[65000:], y1[65000:]
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

```python
model1.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=10, callbacks=[cp1])
```

```python
p_training_mean3 = np.mean(X3_train[:, :, 0])
p_training_std3 = np.std(X3_train[:, :, 0])

temp_training_mean3 = np.mean(X3_train[:, :, 1])
temp_training_std3 = np.std(X3_train[:, :, 1])

def preprocess3(X):
  X[:, :, 0] = (X[:, :, 0] - p_training_mean3) / p_training_std3
  X[:, :, 1] = (X[:, :, 1] - temp_training_mean3) / temp_training_std3
  return X

def preprocess_output(y):
  y[:, 0] = (y[:, 0] - p_training_mean3) / p_training_std3
  y[:, 1] = (y[:, 1] - temp_training_mean3) / temp_training_std3
  return y
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Post-process functions
def postprocess_temp(arr):
    arr = (arr * temp_training_std3) + temp_training_mean3
    return arr

def postprocess_p(arr):
    arr = (arr * p_training_std3) + p_training_mean3
    return arr

# Evaluation function for MSE, RMSE, and R²
def evaluate_model(model, X_test, y_test):
    predictions = model.predict(X_test)

    # Post-process predictions and actual values
    p_preds = postprocess_p(predictions[:, 0])
    temp_preds = postprocess_temp(predictions[:, 1])
    p_actuals = postprocess_p(y_test[:, 0])
    temp_actuals = postprocess_temp(y_test[:, 1])

    # Calculate MSE and RMSE for temperature and pressure
    mse_temp = mean_squared_error(temp_actuals, temp_preds)
    rmse_temp = np.sqrt(mse_temp)

    mse_pressure = mean_squared_error(p_actuals, p_preds)
    rmse_pressure = np.sqrt(mse_pressure)

    # Calculate R² for temperature and pressure
    r2_temp = r2_score(temp_actuals, temp_preds)
    r2_pressure = r2_score(p_actuals, p_preds)

    # Return the results
    return mse_temp, rmse_temp, mse_pressure, rmse_pressure, r2_temp, r2_pressure
```

```python
# Model evaluation for Model 5 and Model 6

model_results = {}

# Evaluate model5
mse_temp5, rmse_temp5, mse_pressure5, rmse_pressure5, r2_temp5, r2_pressure5 = evaluate_model(model5, X3_test, y3_test)
model_results['Model5'] = {
    'MSE Temperature': mse_temp5, 'RMSE Temperature': rmse_temp5,
    'MSE Pressure': mse_pressure5, 'RMSE Pressure': rmse_pressure5,
    'R² Temperature': r2_temp5, 'R² Pressure': r2_pressure5
}

# Evaluate model6
mse_temp6, rmse_temp6, mse_pressure6, rmse_pressure6, r2_temp6, r2_pressure6 = evaluate_model(model6, X3_test, y3_test)
model_results['Model6'] = {
    'MSE Temperature': mse_temp6, 'RMSE Temperature': rmse_temp6,
    'MSE Pressure': mse_pressure6, 'RMSE Pressure': rmse_pressure6,
    'R² Temperature': r2_temp6, 'R² Pressure': r2_pressure6
}
```

```python
# Convert model results to a DataFrame for easy comparison
model_comparison_df = pd.DataFrame(model_results).T
print(model_comparison_df)
```
