```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional, ConvLSTM2D, Flatten, TimeDistributed, Conv1D, MaxPooling1D
from tensorflow.keras.layers import Reshape


# Load dataset
file_path = pd.read_excel('/content/AirQualityUCI.xlsx')

# Preprocessing
def preprocess_data(data, cols, time_steps=10):
    data = data[cols].dropna()
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)

    X, y = [], []
    for i in range(time_steps, len(data_scaled)):
        X.append(data_scaled[i-time_steps:i])
        y.append(data_scaled[i])

    X, y = np.array(X), np.array(y)
    return X, y, scaler

# Extract relevant columns
cols = ['CO(GT)', 'C6H6(GT)']
time_steps = 10
X, y, scaler = preprocess_data(file_path, cols, time_steps)

# Train-test split
split_idx = int(0.8 * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# Function to build and train models
def build_and_train_model(model_type):
    model = Sequential()

    if model_type == 'vanilla_lstm':
        model.add(LSTM(50, input_shape=(time_steps, len(cols))))
    elif model_type == 'stacked_lstm':
        model.add(LSTM(50, return_sequences=True, input_shape=(time_steps, len(cols))))
        model.add(LSTM(50))
    elif model_type == 'bidirectional_lstm':
        model.add(Bidirectional(LSTM(50), input_shape=(time_steps, len(cols))))
    elif model_type == 'cnn_lstm':
        model.add(Conv1D(64, kernel_size=2, activation='relu', input_shape=(time_steps, len(cols))))
        model.add(MaxPooling1D(pool_size=2))
        model.add(LSTM(50))
    elif model_type == 'convlstm':
        model.add(ConvLSTM2D(64, (1, 2), activation='relu', input_shape=(1, time_steps, len(cols), 1)))
        model.add(Flatten())

    model.add(Dense(len(cols)))

    # Compile model
    model.compile(optimizer='adam', loss='mse')

    # Reshape data for ConvLSTM
    if model_type == 'convlstm':
        X_train_reshaped = X_train.reshape((X_train.shape[0], 1, time_steps, len(cols), 1))
        X_test_reshaped = X_test.reshape((X_test.shape[0], 1, time_steps, len(cols), 1))
    else:
        X_train_reshaped, X_test_reshaped = X_train, X_test

    # Train model
    model.fit(X_train_reshaped, y_train, epochs=20, batch_size=32, verbose=1, validation_data=(X_test_reshaped, y_test))
    return model

# Evaluate models
def evaluate_model(model, model_type):
    if model_type == 'convlstm':
        X_test_reshaped = X_test.reshape((X_test.shape[0], 1, time_steps, len(cols), 1))
    else:
        X_test_reshaped = X_test

    predictions = model.predict(X_test_reshaped)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mae = mean_absolute_error(y_test, predictions)

    # Handle division errors for accuracy calculation
    non_zero_indices = y_test != 0
    accuracy = 100 - np.mean(np.abs((y_test[non_zero_indices] - predictions[non_zero_indices]) / y_test[non_zero_indices])) * 100

    # Reverse scaling for predictions
    y_test_scaled = scaler.inverse_transform(y_test)
    predictions_scaled = scaler.inverse_transform(predictions)

    # Display results
    print(f'\nModel: {model_type}')
    print(f'RMSE: {rmse:.3f}')
    print(f'MAE: {mae:.3f}')
    print(f'Accuracy: {accuracy:.2f}%')
    print('Predictions vs Actual Values:')
    for i in range(5):  # Display the first 5 predictions for brevity
        print(f'Actual: {y_test_scaled[i]}, Predicted: {predictions_scaled[i]}')

    plt.figure(figsize=(10, 5))
    plt.plot(y_test_scaled[:, 0], label='Actual CO(GT)')
    plt.plot(predictions_scaled[:, 0], label='Predicted CO(GT)')
    plt.title(f'Model: {model_type} - CO(GT)')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(y_test_scaled[:, 1], label='Actual C6H6(GT)')
    plt.plot(predictions_scaled[:, 1], label='Predicted C6H6(GT)')
    plt.title(f'Model: {model_type} - C6H6(GT)')
    plt.legend()
    plt.show()

# Train and evaluate all models
model_types = ['vanilla_lstm', 'stacked_lstm', 'bidirectional_lstm', 'cnn_lstm', 'convlstm']
for model_type in model_types:
    print(f'\nTraining {model_type}...')
    model = build_and_train_model(model_type)
    evaluate_model(model, model_type)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Multivariate LSTM Models
# Multiple Input Series
def prepare_multiple_input_series(file_path, input_cols, output_cols, time_steps=10):
    file_path = file_path[input_cols + output_cols].dropna()
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(file_path)

    X, y = [], []
    for i in range(time_steps, len(data_scaled)):
        X.append(data_scaled[i-time_steps:i, :len(input_cols)])
        y.append(data_scaled[i, len(input_cols):])

    X, y = np.array(X), np.array(y)
    return X, y, scaler

# Prepare data for multiple input series
input_cols = ['CO(GT)', 'C6H6(GT)']
output_cols = ['CO(GT)', 'C6H6(GT)']
X_multi_input, y_multi_input, scaler_multi_input = prepare_multiple_input_series(file_path, input_cols, output_cols, time_steps)

# Build and Train Model for Multiple Input Series
def build_multi_input_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, input_shape=input_shape))
    model.add(Dense(len(output_cols)))
    model.compile(optimizer='adam', loss='mse')
    return model

multi_input_model = build_multi_input_model((time_steps, len(input_cols)))
multi_input_model.fit(X_multi_input, y_multi_input, epochs=20, batch_size=32, verbose=1)

# Evaluate Model for Multiple Input Series
def evaluate_multi_input_model(model, X_test, y_test, scaler):
    predictions = model.predict(X_test)

    # Reshape predictions to have the same number of features as the scaler
    predictions_reshaped = np.zeros((predictions.shape[0], scaler.n_features_in_))
    predictions_reshaped[:, len(input_cols):len(input_cols) + predictions.shape[1]] = predictions  # Adjust the slicing

    # Reshape y_test to have the same number of features as the scaler
    y_test_reshaped = np.zeros((y_test.shape[0], scaler.n_features_in_))
    y_test_reshaped[:, len(input_cols):len(input_cols) + y_test.shape[1]] = y_test  # Adjust the slicing

    y_test_scaled = scaler.inverse_transform(y_test_reshaped)[:, len(input_cols):len(input_cols) + y_test.shape[1]] # Select the correct columns after inverse_transform
    predictions_scaled = scaler.inverse_transform(predictions_reshaped)[:, len(input_cols):len(input_cols) + predictions.shape[1]]  # Inverse transform the reshaped predictions and select correct cols

    rmse = np.sqrt(mean_squared_error(y_test_scaled, predictions_scaled))
    mae = mean_absolute_error(y_test_scaled, predictions_scaled)

    # Calculate R-squared
    r2 = r2_score(y_test_scaled, predictions_scaled)

    # Calculate R-squared Percentage  <-- Added this line
    r2_percentage = r2 * 100

    # Print RMSE, MAE, and R-squared Percentage  <-- Modified this line
    print(f'RMSE: {rmse:.3f}, MAE: {mae:.3f}, R-squared: {r2:.3f} ({r2_percentage:.2f}%)')

    plt.figure(figsize=(10, 5))
    plt.plot(y_test_scaled[:, 0], label='Actual CO(GT)')
    plt.plot(predictions_scaled[:, 0], label='Predicted CO(GT)')
    plt.legend()
    plt.show()

evaluate_multi_input_model(multi_input_model, X_multi_input, y_multi_input, scaler_multi_input)

# Multiple Parallel Series
# For simplicity, using the same input_cols and output_cols
X_parallel, y_parallel, scaler_parallel = prepare_multiple_input_series(file_path, input_cols, output_cols, time_steps)

parallel_model = build_multi_input_model((time_steps, len(input_cols)))
parallel_model.fit(X_parallel, y_parallel, epochs=20, batch_size=32, verbose=1)

evaluate_multi_input_model(parallel_model, X_parallel, y_parallel, scaler_parallel)

# End
```
